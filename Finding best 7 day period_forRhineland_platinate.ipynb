```python
import csv 
import pandas as pd
from datetime import date, timedelta


read_cs = pd.read_csv('1yearDataUpdated_Aug1st2021ToJuly31st2022_Updated_final.csv')
read_cs["Date"] = pd.to_datetime(read_cs["Date"])
```

```python
selected_df =  read_cs[['Date','District','temperature','wind_speed','relative_humidity']]
group_by_date_df = selected_df.groupby(['Date'], as_index=False, sort=False).agg(avg_temperature_per_day = ('temperature', 'mean'), avg_windspeed_per_day = ('wind_speed', 'mean'), avg_relativehumidity_per_day = ('relative_humidity', 'mean'))
group_by_date_df["Date"] = pd.to_datetime(group_by_date_df["Date"])
group_by_date_df
```

```python
min_value = 0
def get_7_days(startdate, enddate, incubation_days, temperature_list_tuple, min_value):
    counter = 1
    sum_temperature = 0
    if counter <= incubation_days:
         for single_date in daterange(startdate, enddate):
                currentdate = single_date.strftime("%d-%m-%Y")
                # print(currentdate)
                df = group_by_date_df[group_by_date_df['Date'] == currentdate]
                #val = df['avg_temperature_per_day'].values[0]
                val = df['avg_temperature_per_day'].values[0]
                #val = df['avg_relativehumidity_per_day'].values[0]
                #print('currentdate',currentdate,'Avg temperature',val,'counter',counter)
                sum_temperature +=val
                if counter ==7 :
                    mean_temperature = sum_temperature/counter
                    temperature_list_tuple.append((str(startdate),currentdate,mean_temperature))
                    #print(temperature_list_tuple)
                counter +=1
               
    

def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
        
start_date = date(2022,6,4) #yyyy-mm-dd
end_date = date(2022,7,30)
incubation_days = 7
temperature_list = []

for single_date in daterange(start_date, end_date):
    get_7_days(single_date, end_date, incubation_days, temperature_list, min_value)
#print(temperature_list)
for index,listtuple in enumerate(temperature_list):
    print(listtuple)
    if listtuple[2] > min_value :
        min_value = listtuple[2]
print("min_value",min_value)
```
